```python
from processes import OUProcess
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_len = 200
n_samples = 3
```

# Ornsteinâ€“Uhlenbeck process

Unmodified Ornstein-Uhlenback process.
There's a tendency to return to the origin, but no memory.

```python
ou_process = OUProcess(decay_theta=0.02, sigma=1.0, initial_value=np.zeros(dtype=np.float32, shape=(1,)))
ou_samples = []
for _ in range(n_samples):
    ou_process.reset()
    ou_samples.append([ou_process.sample() for _ in range(sample_len)])
```

```python
for sample in ou_samples:
    plt.plot(sample, alpha=0.75)
plt.show()
```
