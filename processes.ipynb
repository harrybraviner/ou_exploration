```python
from processes import OUProcess, OUProcessWithMemory
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_len = 200
n_samples = 3
```

# Ornstein–Uhlenbeck process

Unmodified Ornstein-Uhlenback process.
There's a tendency to return to the origin, but no memory.

```python
ou_process = OUProcess(decay_theta=0.02, sigma=1.0, initial_value=np.zeros(dtype=np.float32, shape=(1,)))
ou_samples = []
for _ in range(n_samples):
    ou_process.reset()
    ou_samples.append([ou_process.sample() for _ in range(sample_len)])
```

```python
for sample in ou_samples:
    plt.plot(sample, alpha=0.75)
plt.show()
```

# Ornstein–Uhlenbeck process with memory

Modified Ornstein-Uhlenback process in which the process is attracted towards recent states.

```python
ou_process = OUProcessWithMemory(decay_theta=0.02, sigma=1.0, initial_value=np.zeros(dtype=np.float32, shape=(1,)))
ou_samples = []
for _ in range(n_samples):
    ou_process.reset()
    ou_samples.append([ou_process.sample() for _ in range(sample_len)])
```

```python
for sample in ou_samples:
    plt.plot(sample, alpha=0.75)
plt.show()
```

Now what if we don't make this attracted towards the recent average, but the negative of this?
(This is caused by the `av_sign=-1.0` argument in the `OUProcessWithMemory` constructor.)

```python
ou_process = OUProcessWithMemory(decay_theta=0.02, sigma=1.0, initial_value=np.zeros(dtype=np.float32, shape=(1,)),
                                av_sign=-1.0)
ou_samples = []
for _ in range(n_samples):
    ou_process.reset()
    ou_samples.append([ou_process.sample() for _ in range(sample_len)])
```

```python
for sample in ou_samples:
    plt.plot(sample, alpha=0.75)
plt.show()
```
